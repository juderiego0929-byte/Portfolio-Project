# Data Cleaning in Pandas 

```python
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\ACER\Documents\DATA ANALYST\PANDAS\8. Data Cleaning\Customer Call List.xlsx")

df
```

```python
# remove Duplicateas
# always save after deleting duplicates or after editing, use the same variable to save it

df = df.drop_duplicates()

df
```

```python
# deleting unnecessary columns

df = df.drop(columns = "Not_Useful_Column")
```

```python
df
```

```python
# strip is the position of outer left and outer right of the value, its like the first and last letter
# you can also use lstrip if you want to edit only the left strip or first letter and you can also use rstrip if you want to edit the last letter
# specify the column that you want to edit
# removing unnecessary letters or special characters
# always call the columns that you want to edit if you want to save it already

# first way
#df["Last_Name"] = df["Last_Name"].str.lstrip("/")
#df["Last_Name"] = df["Last_Name"].str.lstrip("...")
#df["Last_Name"] = df["Last_Name"].str.rstrip("_")

# second way
# you can put all the strings that you want to remove in the same code
df["Last_Name"] = df["Last_Name"].str.strip("/._")

df
```

```python
# removing the special characters
# its gonna return any characters except not a-zA-Z0-9 and replace it in to none or blank or ""
# ^ - this symbol is  = not
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','', regex = True)

# making the float into a string
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))
# standardize the phone number 
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])

# Removing the nan-- and Na--
df["Phone_Number"] = df["Phone_Number"].str.replace("nan--", "")
df["Phone_Number"] = df["Phone_Number"].str.replace("Na--", "")
df["Phone_Number"] = df["Phone_Number"].str.replace("--", "")

df
```

```python
# Separating columns in Address each Street address, State, zipcode
df[["Street Address", "State", "Zip Code"]]=df["Address"].str.split(',',n = 2, expand = True)

df
```

```python
#editing the n and yes value in Paying Customer

df["Paying Customer"] = df["Paying Customer"].str.replace("Yes", "Y")
df["Paying Customer"] = df["Paying Customer"].str.replace("Y", "Yes")
df["Paying Customer"] = df["Paying Customer"].str.replace("No", "N")
df["Paying Customer"] = df["Paying Customer"].str.replace("N", "No")
df["Paying Customer"] = df["Paying Customer"].str.replace("No/a", "")
df
```

```python
#editing the n and yes value in do_not_contact

df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace("Yes", "Y")
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace("Y", "Yes")
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace("No", "N")
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace("N", "No")
df
```

```python
# replacing the N/a into a blank
df = df.replace("N/a", "")

#removing all none
df = df.fillna("")
```

```python
df
```

```python
# removing the person that dont want to contact
for w in df.index:
    if df.loc[w,"Do_Not_Contact"] == "Yes":
        df.drop(w, inplace = True)

df

#Another way to drop null values
#df = df.dropna(subset="Phone_Number"), inplace=True)
```

```python
#removing the blank phone number

for w in df.index:
    if df.loc[w,"Phone_Number"] == "":
        df.drop(w, inplace = True)

df
```

```python
# reseting the index
df.reset_index(drop = True)
```
